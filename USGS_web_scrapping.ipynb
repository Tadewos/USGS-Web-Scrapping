```python
#html_url = "https://waterdata.usgs.gov/nwis/uv?cb_all_=on&cb_00045=on&format=rdb&site_no=411607084241201&period=&begin_date=2021-11-15&end_date=2022-03-15"
'''
html_url = "https://waterdata.usgs.gov/nwis/uv?cb_all_=on&cb_00045=on&format=html&site_no=411607084241201&period=&begin_date=2021-11-15&end_date=2022-03-15"
html_url2 = "https://waterdata.usgs.gov/nwis/dv?cb_00060=on&format=rdb&site_no=411607084241200&referred_module=sw&period=&begin_date=2021-06-16&end_date=2022-03-28"
html_url3 = "https://nwis.waterdata.usgs.gov/nwis/dv?cb_00060=on&format=html&site_no=411610084240800&referred_module=sw&period=&begin_date=2021-05-22&end_date=2022-03-14"
'''
#Work with Tab Separated Data instead
html_hist1 = 'https://nwis.waterdata.usgs.gov/usa/nwis/uv/?cb_00045=on&format=rdb&site_no=411607084241201&period=&begin_date=2021-12-01&end_date=2022-03-31'
html_hist2 = 'https://nwis.waterdata.usgs.gov/usa/nwis/uv/?cb_00060=on&format=rdb&site_no=411607084241200&period=&begin_date=2021-06-15&end_date=2022-03-29'
html_daily2 = 'https://waterdata.usgs.gov/nwis/dv?cb_00060=on&format=rdb&site_no=411607084241200&referred_module=sw&period=&begin_date=2021-06-16&end_date=2022-03-28'
html_hist3 = 'https://nwis.waterdata.usgs.gov/nwis/uv?cb_00060=on&format=rdb&site_no=411610084240800&period=&begin_date=2021-05-21&end_date=2022-03-29'
html_daily3 = 'https://nwis.waterdata.usgs.gov/nwis/dv?cb_00060=on&format=rdb&site_no=411610084240800&referred_module=sw&period=&begin_date=2021-03-28&end_date=2022-03-28'
```

Process Data

```python
precip = pd.read_csv(html_hist1, header=27, sep='\t', parse_dates=['20d'])
discharge1 = pd.read_csv(html_hist2, header=27, sep='\t', parse_dates=['20d'])
mean_discharge1 = pd.read_csv(html_daily2, header=28, sep='\t', parse_dates=['20d'])
discharge2 = pd.read_csv(html_hist3, header=28, sep='\t', parse_dates=['20d'])
mean_discharge2 = pd.read_csv(html_daily3, header=29, sep='\t', parse_dates=['20d'])
```
